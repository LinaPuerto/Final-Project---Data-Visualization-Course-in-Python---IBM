Una forma de hacerlo es - usar as_index como falso, de lo contrario terminars con un dataframe con mltiples ndices. ````python

````python
df_rec = df[df['Recession'] == 1]
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

df_pivot = df_Mline.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')
df_pivot.plot(kind='line', marker='o', figsize=(12, 6))

plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Sales Trend Vehicle-wise during Recession')

plt.xticks(ticks=df_pivot.index, labels=df_pivot.index, rotation=65)

# Lneas verticales para aos de recesin clave
for year in [2007, 2009, 2020]:
    if year in df_pivot.index:
        plt.axvline(x=year, color='gray', linestyle='--', linewidth=1.5)

plt.grid(True)
plt.tight_layout()
plt.legend(title='Vehicle Type')
plt.show()
````

````python
# Filter the data for recessionary periods
#Agrupar por tipos de vehculos y estado de recesin
grouped = df.groupby(['Vehicle_Type', 'Recession'])['Automobile_Sales'].mean().reset_index()

# Map recession flag to labels
grouped['Recession'] = grouped['Recession'].map({0: 'Non-Recession', 1: 'Recession'})
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped, x='Vehicle_Type', y='Automobile_Sales', hue='Recession',
            palette={'Non-Recession': '#53ad81', 'Recession': '#c4674c'})

# Customize labels and title
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.title('Comparison of Vehicle Sales: Recession vs Non-Recession')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
#impacto de la estacionalidad en las ventas de automviles Cmo ha afectado la estacionalidad a las ventas?
#En qu meses fueron altas o bajas? Revselo para aos sin recesin para comprender la tendencia.

# Filtrar solo datos de aos sin recesin
non_rec_data = df[df['Recession'] == 0]

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=non_rec_data,
    x='Month',  # Eje X: Mes
    y='Automobile_Sales',  # Eje Y: Ventas
    size='Seasonality_Weight',  # Tamao de burbuja
    hue='Seasonality_Weight',  # Color opcional segn peso estacional
    sizes=(50, 500),  # Tamao mnimo y mximo de burbuja
    legend=False,
    alpha=0.3)

plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Impact on Automobile Sales (Non-Recession Years)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import folium
import requests

url_csv = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url_csv)
df.columns = df.columns.str.strip()  # Limpiar nombres

# Filtrar solo aos en recesin
df_rec = df[df['Recession'] == 1]

sales_by_state = df_rec.groupby("City")["Automobile_Sales"].sum().reset_index()
sales_by_state.columns = ["State", "Total_Sales"]

geo_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/us-states.json"
geo_response = requests.get(geo_url)
us_states_geo = geo_response.json()

m = folium.Map(location=[37.8, -96], zoom_start=4)

choropleth = folium.Choropleth(
    geo_data=us_states_geo,
    data=sales_by_state,
    columns=['State', 'Total_Sales'],
    key_on='feature.id',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Total Automobile Sales During Recession',
    highlight=True
).add_to(m)


folium.GeoJson(
    us_states_geo,
    name="Labels",
    style_function=lambda x: {'fillColor': 'transparent', 'color': 'transparent', 'weight': 0},
    tooltip=folium.GeoJsonTooltip(
        fields=['name'],
        aliases=['State:'],
        localize=True
    )
).add_to(m)

# Mostrar el mapa
m

````
